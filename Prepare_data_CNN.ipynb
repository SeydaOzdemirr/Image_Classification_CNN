# Open the tar.gz folder

```python
import tarfile
import os
path = "../case"
```

```python
os.chdir(path)
fname = os.listdir(".")[1]
fname 
```

```python
if fname.endswith("tar.gz"):
    tar = tarfile.open(fname,"r:gz")
    tar.extractall()
    tar.close()
elif fname.endswith("tar"):
    tar = tarfile.open(fname,"r:")
    tar.extractall()
    tar.close()
```
